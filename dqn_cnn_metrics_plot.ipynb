```python
import plotly.express as px
import pandas as pd
```

```python
df_dqncnn = pd.read_parquet('output/20240603_DDQAugmentedTransformerNN_metrics.pq')
```

```python
df_dqncnn
```

```python
df_dqncnn_melted = pd.melt(df_dqncnn, id_vars=['episode'], var_name='value_type', value_name='value').sort_values(['episode', 'value_type']).reset_index(drop=True)
```

```python
fig = px.line(df_dqncnn_melted[df_dqncnn_melted['value_type'] == 'epsilon'], x="episode", y="value", color='value_type', title='Epsilon Decay per Episode')
fig.show()
```

```python
fig = px.line(df_dqncnn_melted[df_dqncnn_melted['value_type'].isin(['avg_episode_score', 'avg_episode_loss'])], x="episode", y="value", color='value_type', title='Loss/Score per Episode')
fig.show()
```

```python
fig = px.line(df_dqncnn_melted[df_dqncnn_melted['value_type'].isin(['count_predicted_actions', 'count_random_action'])], x="episode", y="value", color='value_type', title='Predicted vs. Random Actions per Episode')
fig.show()
```
